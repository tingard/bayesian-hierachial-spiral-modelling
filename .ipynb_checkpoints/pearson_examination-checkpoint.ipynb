Sandbox playing with pearson correlations

```python
import scipy.stats as st
```

```python
def gen_data(N, a=1):
    x = st.norm.rvs(size=N)
    sigma = st.norm.rvs(size=N)
    return x, a * x + sigma
```

```python
A = 0.02
plt.figure(figsize=(8, 8), dpi=60)
plt.hexbin(*gen_data(100000, a=A));
```

By eye, we would say there is next to no correlation.

```python
N_low = 100
df_low = pd.DataFrame(
    [st.pearsonr(*gen_data(N_low, a=A)) for _ in range(500)],
    columns=('statistic', 'p_value'),
)
df_low.agg(['mean', 'std', 'min', 'max']).T
```

The test with $N = 100$ agrees, with a large variation in outcomes (including massive variations in statistical significance)

```python
N_med = 10000
df_med = pd.DataFrame(
    [st.pearsonr(*gen_data(N_med, a=A)) for _ in range(100)],
    columns=('statistic', 'p_value'),
)
df_med.agg(['mean', 'std', 'min', 'max']).T
```

$N = 10,000$ is more consistent, though again with a large scatter on p-values

```python
N_high = 1000000
df_high = pd.DataFrame(
    [st.pearsonr(*gen_data(N_high, a=A)) for _ in range(100)],
    columns=('statistic', 'p_value'),
)
df_high.agg(['mean', 'std', 'min', 'max']).T
```

$N = 1,000,000$ is where consistency starts to kick in: the results are always significant and, as expected, state there is no correlation.
